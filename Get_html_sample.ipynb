Note: Digikey has forbidden access to the data. API access is needed for that. So, we use maker's website.

```
from bs4 import BeautifulSoup
from urllib.request import urlopen
url = 'https://www.freecollocation.com/'
search_word = 'search?word=talk'

html = urlopen(url+search_word)
soup = BeautifulSoup(html)#Soup is better to read than urlopen.
#print(html.read())
print(soup)
```
